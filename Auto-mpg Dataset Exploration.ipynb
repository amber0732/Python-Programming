```python
## Median :-
median1 = df["mpg"].median()
median1
```

### Droping Car Name Column :-

```python
df.drop("car name", axis = 1, inplace = True)
```

### Dealing with the missing values :-

```python
df["horsepower"] = df["horsepower"].replace("?", np.nan)
```

```python
df["horsepower"] = df["horsepower"].astype(float)
```

```python
median1 = df["horsepower"].median()
median1
```

```python
df["horsepower"].replace(np.nan, median1, inplace = True)
```

```python
df.dtypes
```

### Replacing Origin with Country names :-

```python
df["origin"] = df["origin"].replace({1 : "India", 2 : "Korea", 3 : "America"})
```

```python
df.sample(10)
```

## Encoding :-
### -> It is only done by using pandas Library.

```python
df = pd.get_dummies(df, columns = ["origin"])
```
